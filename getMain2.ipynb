```python
# Initialize an empty list to store the results
article_list = []

# Loop through each row in the DataFrame using iterrows()
for index, row in df.iterrows():
    try:
        Link = row['Link']  # Access the 'Link' column by name
        Title = row['Title']  # Access the 'Title' column by name
        Time_Uploaded = row['Time Uploaded']  # Access the 'Time Uploaded' column by name
        Author = row['Author']  # Access the 'Author' column by name
        Tags = row['Tags']  # Access the 'Tags' column by name (check if there’s an extra space)
        Reading_Time = row['Reading Time']  # Access the 'Reading Time' column by name
        Article_Content = row['Article Content']  # Access the 'Article Content' column by name
        Word_Count = row['Word Count']  # Access the 'Word Count' column by name
        Sentiment = row['Sentiment']  # Access the 'Sentiment' column by name
        Compound_Score=row['Compound Score']
    except KeyError as e:
        print(f"Column not found: {e}")
        continue  # Skip rows where any column is missing
    
    # Append the tuple (Link, Title, Time_Uploaded, Author, Tags, Reading_Time, Article_Content, Word_Count, Sentiment)
    article_list.append((Link, Title, Time_Uploaded, Author, Tags, Reading_Time, Article_Content, Word_Count, Sentiment,Compound_Score))

# Now article_list contains tuples of (Link, Title, Time_Uploaded, Author, Tags, Reading_Time, Article_Content, Word_Count, Sentiment)
df_list = pd.DataFrame(article_list, columns=["Link", "Title", "Time_Uploaded", "Author", "Tags", "Reading_Time", "Article_Content", "Word_Count", "Sentiment","Compound_Score"])

df_list  # Display the new DataFrame
```

```python
# Convert column names to lowercase
df_list.columns = df_list.columns.str.lower()
df_list
```

```python
#Access the "reading_time" column.
#For each value in the column, remove the string " min read" from the end.
#Ensure that the remaining value in the column is converted to the appropriate datatype, probably an integer.
# Convert 'reading_time' column to string type first
df_list['reading_time'] = df_list['reading_time'].astype(str)

# Remove the string " min read" from the "reading_time" column
df_list['reading_time'] = df_list['reading_time'].str.replace(' min read', '', regex=False)

# Convert the remaining values in the "reading_time" column to integers
df_list['reading_time'] = df_list['reading_time'].astype(int)

# Display the updated DataFrame
df_list
```

```python
#Access the "article_content" column.
#For each value in the column, replace all occurrences of the newline character (\n) with an empty string ("").

df_list['article_content'] = df_list['article_content'].str.replace('\n', '', regex=False)

# Display the updated DataFrame
df_list
```

```python
#After the transformations, ensure that each column has the correct data type (e.g., integer, string, date, etc.).
# You might need to use Pandas functions like astype() to enforce these data types.
#Determine the appropriate data types for each column based on the data that it contains.
# Ensure columns have the correct data type after transformations

# Convert 'reading_time' to integer (after removing 'min read')
df_list['reading_time'] = df_list['reading_time'].astype(int)

# Convert 'compound_score' to float (since it's likely a score or rating)
df_list['compound_score'] = df_list['compound_score'].astype(float)

# Convert 'time_uploaded' to datetime (assuming it's in a recognizable date format)
df_list['time_uploaded'] = pd.to_datetime(df_list['time_uploaded'], errors='coerce')

# Convert 'link', 'title', 'author', 'tags', and 'article_content' to strings
df_list['link'] = df_list['link'].astype(str)
df_list['title'] = df_list['title'].astype(str)
df_list['author'] = df_list['author'].astype(str)
df_list['tags'] = df_list['tags'].astype(str)
df_list['article_content'] = df_list['article_content'].astype(str)

# Check the data types of each column to verify
print(df_list.dtypes)
```
